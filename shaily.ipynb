```python
df = pd.read_csv("DSC209_FinalProject/data/Indicator_1_1_annual_6562429754166382300.csv")
```

```python
df.head()
```

```python
years = [str(y) for y in range(2010, 2024)]

world_ghg = df[
    (df["Country"] == "World") &
    (df["Gas Type"] == "Greenhouse gas")
]
```

```python
tidy = world_ghg.melt(
    id_vars=["Industry"],
    value_vars=years,
    var_name="Year",
    value_name="Emissions"
)

tidy = tidy.sort_values(["Year", "Emissions"], ascending=[True, False])
tidy.head()
```

```python
output_path = "DSC209_FinalProject/data/world_ghg_by_industry.csv"
tidy.to_csv(output_path, index=False)
output_path
```
